```python
import pandas as pd
import os

# Step 1: Check current directory
print("Current Directory:", os.getcwd())

# Step 2: List files to confirm CSV exists
print("Files in directory:", os.listdir())

file_path = "Admission_Chance.csv"

# Step 3: Check if file exists
if not os.path.exists(file_path):
    print("❌ File not found. Check the file path.")
else:
    try:
        # Step 4: Try reading normally
        df = pd.read_csv(file_path)
        
        # Step 5: Check if dataframe is empty
        if df.empty:
            print("⚠️ File is empty.")
        else:
            print("✅ File loaded successfully!")
            print(df.head())

    except pd.errors.EmptyDataError:
        print("❌ File exists but contains no data.")

    except pd.errors.ParserError:
        print("⚠️ Parsing issue. Trying different separator...")
        try:
            df = pd.read_csv(file_path, sep=';')
            print("✅ Loaded with semicolon separator")
            print(df.head())
        except:
            print("❌ Could not parse file.")

    except Exception as e:
        print("❌ Unexpected error:", e)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Example dataset
df = pd.DataFrame({
    "a": [1, 2, 3, 4],
    "b": [4, 3, 2, 1],
    "target": [10, 20, 30, 40]
})

X = df.drop("target", axis=1)
y = df["target"]

rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X, y)

import matplotlib.pyplot as plt

plt.barh(X.columns, rf_model.feature_importances_)
plt.show()
```
