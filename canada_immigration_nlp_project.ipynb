```python
import pandas as pd

df = pd.read_csv('Canada Immigration.csv')
X = df.iloc[:, :-1].values  # all columns except last
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)

df['Cluster'] = kmeans.fit_predict(X)

df.head()
```

```python
col = 'Immigration_Reason'  # update if needed

for i in range(3):
    print(f"\nCluster {i} Sample Reasons:")
    
    subset = df[df['Cluster'] == i]
    
    if col in df.columns:
        print(subset[col].head(5))
    else:
        print(f"Column '{col}' not found!")
```
