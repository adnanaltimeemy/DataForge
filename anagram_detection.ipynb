```python
def extract_features(row):
    w1 = row['word1']
    w2 = row['word2']
    
    # Length difference
    length_diff = abs(len(w1) - len(w2))
    
    # Character frequency difference
    freq1 = [w1.count(chr(i)) for i in range(97, 123)]
    freq2 = [w2.count(chr(i)) for i in range(97, 123)]
    
    freq_diff = np.sum(np.abs(np.array(freq1) - np.array(freq2)))
    
    return pd.Series([length_diff, freq_diff])
```
