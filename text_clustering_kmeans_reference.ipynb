```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans

# Example text data (2 documents)
documents = ["text data here", "more text data"]

# Step 1: Fit vectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# Step 2: Fit KMeans (k must be <= number of documents)
k = 2
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
kmeans.fit(X)

# Step 3: Extract feature names
terms = vectorizer.get_feature_names_out()

# Step 4: Print top words per cluster
for i in range(k):
    center_terms = kmeans.cluster_centers_[i].argsort()[-10:][::-1]
    print(f"\nCluster {i} top words:")
    print([terms[ind] for ind in center_terms])
```
